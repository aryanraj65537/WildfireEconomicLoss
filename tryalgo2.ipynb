```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv("output.csv")
```

```python
data.columns = ["Elevation", "Temp", "Biomass", "Fire"]
```

```python
data
```

```python
import random

data.replace(('True', 'False'), (1, 0), inplace=True)

for x in range(len(data.Fire)):
    if data.Fire[x] == False and random.random()<0.85:
        data = data.drop(labels=x, axis=0)
data
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=99)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(criterion='gini', max_depth=8, min_samples_split=10, random_state=3)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, x_train, y_train, cv=10)
```
